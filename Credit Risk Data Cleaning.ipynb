```python
# Dealing with nulls in person emp length
before = df['person_emp_length'].isnull().sum()
df['person_emp_length'] = np.where(
    df['person_emp_length'].isnull(), df['person_age']- 16, df['person_emp_length']
)
after = df['person_emp_length'].isnull().sum()

print(f"Nulls filled:{before - after}")
```

```python
# Dealing with logical inconsistency where emp length exceeds age-16
# Assuming min age to bein working is 16
df['was_impossible'] = (df['person_emp_length'] > (df['person_age']- 16))

df['person_emp_length']= np.where(
    (df['person_emp_length'] > (df['person_age']- 16)), df['person_age']- 16,
    df['person_emp_length']
)

corrected_rows = df[df['was_impossible']]
print(f"Total Corrected: {len(corrected_rows)}")
print(corrected_rows[['person_age', 'person_emp_length', 'was_impossible']])

```

```python
# Dealing with nulls in Loan interest rate
# Filling the nulls with the median of the loan grade

before = df['loan_int_rate'].isnull().sum()
df['loan_int_rate']= df.groupby('loan_grade')['loan_int_rate']\
                        .transform(lambda x: x.fillna(x.median()))

after = df['loan_int_rate'].isnull().sum()
print(f"Nulls for int rate filled: {before - after}")

```

```python
# Checking & Solving logical inconsistencies
# Age - Should not be anyone below 18 and capping it at 80
print(f"Age < 18: {(df['person_age'] < 18).sum()}")
print(f"Age > 80: {(df['person_age'] > 80).sum()}")
```

```python
before = df.shape[0]

df = df[df['person_age'] < 80]

after = df.shape[0]

print(f"Rows removed: {before - after}")
```
